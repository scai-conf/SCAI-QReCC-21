```python
# load dataset
import json

path = '../../dataset/qrecc_test.json'

with open(path) as f: 
    dialogues = json.load(f)

    
path = '../../dataset/scai-qrecc21-ground-truth.json'

with open(path) as f: 
    gt = json.load(f)
```

```python
# show example
print(dialogues[0])
print(gt[0])
```

```python
print(len(dialogues), 'samples')
print(len(gt), 'annotations')
assert len(dialogues) == len(gt)
```

```python
# count missing gt passages
n_empty_answers = 0 
for sample in gt:
    if not sample['Truth_passages']:
        n_empty_answers += 1
print(n_empty_answers, 'questions without answer passages')
print('out of %d questions' % len(gt))
```
