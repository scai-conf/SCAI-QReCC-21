```python
# load annotations
import pandas as pd

df = pd.read_csv('../annotations/generated-answers-faithfulness.csv')
# print(df.head())
print(len(df), 'rows')
```

```python
# replace the radio button results with a single column
df.loc[df['Answer.irrelevant.on'] == True, 'Annotation'] = 'Irrelevant'
df.loc[df['Answer.nosupport.on'] == True, 'Annotation'] = 'Unfaithful'
df.loc[df['Answer.support.on'] == True, 'Annotation'] = 'Faithful'
# drop old columns
df = df.drop(['Answer.irrelevant.on', 'Answer.nosupport.on', 'Answer.support.on', 'Approve', 'Reject'], axis = 1)
df
```

```python
# remove all duplicate rows where the judgments agreed for the same id, answer and annotation
df = df.drop_duplicates(subset=['Input.id', 'Input.answer', 'Annotation'], keep='last')
len(df)
```

```python
# save resulting table with disagreements that were not resolved manually by the quality evaluator
df.to_csv('../annotations/answer_faithfulness_annotations_clean_with_disagreements.csv')
```

```python
# remove all rows with disagreements
# ie all duplicate rows on the first two columns that were not removed at the previous step
df = df.drop_duplicates(subset=['Input.id', 'Input.answer'], keep=False)
print(len(df), 'answers on which annotators agreed')
```

```python
df_unfaithful = df[df['Annotation']=='Unfaithful']
print(len(df_unfaithful), 'unfaithful answers')
df_unfaithful = df_unfaithful.drop(['Annotation'], axis = 1)
# save resulting table with disagreements that were not resolved manually by the quality evaluator
df_unfaithful.to_csv('../annotations/unfaithful_answers.csv')
df_unfaithful
```

```python
df_unfaithful.groupby(['Input.model']).size().sort_values(ascending=False)
```

```python
# select the subset with only faithful answers (ie annotated as faithful by all annotators)
df_faithful = df[df['Annotation']=='Faithful']
print(len(df_faithful), 'faithful answers')
df_faithful = df_faithful.drop(['Annotation'], axis = 1)
df_faithful.to_csv('../annotations/faithful_answers.csv')
df_faithful
```

```python
# calculate the number of faithful answers by each model
df_faithful.groupby(['Input.model']).size().sort_values(ascending=False)
```
